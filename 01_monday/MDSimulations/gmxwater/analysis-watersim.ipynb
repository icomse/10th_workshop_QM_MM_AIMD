```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO
```

```python
def read_datafile(filepath):
    clean_lines = []
    with open(filepath, 'r') as f:
        for line in f:
            if not line.startswith('#') and not line.startswith('@'):
                clean_lines.append(line.strip())
    data_io = StringIO('\n'.join(clean_lines))
    df = pd.read_csv(data_io, sep=r'\s+', header=None)
    return df
```

```python
df1 = read_datafile('energy-spce1024long.xvg')
```

```python
# View structure
df1.head()
```

```python
plot_labels = ['Time (ps)','P.Energy (kJ/mol)', 'Total Energy','Temperature (K)', 'Pressure', 'Volume', 'Density']

# Step 4: Plot each Y-column using labels, not df column names
x = df1.iloc[:, 0]

for i in range(1, df1.shape[1]):
    y = df1.iloc[:, i]
    plt.figure(figsize=(6, 4))
    plt.plot(x, y, marker='o')
    plt.xlabel(plot_labels[0])
    plt.ylabel(plot_labels[i])
    plt.title(f'{plot_labels[i]} vs {plot_labels[0]}')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
df2 = read_datafile('rdf-spce1024long.xvg')
df2.head()
```

```python
# Step 4: Plot each Y-column using labels, not df column names
x = df2.iloc[:, 0]
y = df2.iloc[:, 1]
plt.figure(figsize=(6, 4))
plt.plot(x, y, marker='o')
plt.xlabel('Distance')
plt.ylabel('RDF')
plt.grid(True)
plt.tight_layout()
plt.show()
```
